### Quantum Circuits and Gates + Qasm Simulator

```python
# Create a Quantum Circuit acting on the q register Here, you are initializing with 2 qubits in the zero state;
# with 2 classical bits set to zero.
circuit = QuantumCircuit(2, 2)

# Add a H gate on qubit 0
circuit.h(0)

# Add a CX (CNOT) gate on control qubit 0 and target qubit 1
circuit.cx(0, 1)

# Map the quantum measurement to the classical bits
circuit.measure([0, 1], [0, 1])

# Draw the circuit
circuit.draw()
```

```python
circuit.draw('mpl', style='clifford')
```

In this circuit, the qubits are ordered with qubit zero at the top and qubit one at the bottom. The circuit is read left-to-right, meaning that gates which are applied earlier in the circuit show up farther to the left.

It is important to note that gate multiplication reads left to right, whereas matrix multiplication is right to left. ( CNOT*H|00>)

Simulate the Experiment using **qasm simulator**

Bell states: https://en.wikipedia.org/wiki/Bell_state

```python
# Bell state
circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0, 1], [0, 1])
circuit.draw('mpl', style='clifford')
```

```python
# Use Aer's qasm_simulator
# simulator = BasicAer.get_backend('qasm_simulator')
simulator = Aer.get_backend('qasm_simulator')

# Compile the circuit for the support instruction set (basis_gates) and topology (coupling_map) of the backend
compiled_circuit = transpile(circuit, simulator)

Number_of_shots = 2000
# Execute the circuit on the aer simulator
job = simulator.run(compiled_circuit, shots= Number_of_shots)
#The number of shots of the simulation was set to be 2000 (the default is 1024).

# Grab results from the job
result = job.result()
```

```python
job.status()
```

```python
# Returns counts
counts = result.get_counts(compiled_circuit)
print("\nTotal count for 00 and 11 are:", counts)
```

As expected, the output bit string is 00 approximately 50 percent of the time.

```python
#Visualize the Results
plot_histogram(counts)
```

The observed probabilities Pr(00) and Pr(11) are computed by taking the respective counts and dividing by the total number of shots.

```python
# Print the probabilities
for state, count in counts.items():
    probability = count / Number_of_shots
    print(f"State {state}: Probability {probability}")
```

### Statevector backend :
The statevector_simulator executes quantum circuits and returns the final state vector of the quantum system, assuming perfect quantum device (no noise or decoherence).

Greenberger–Horne–Zeilinger state (GHZ state) : https://en.wikipedia.org/wiki/Greenberger%E2%80%93Horne%E2%80%93Zeilinger_state

```python
# Create a Quantum Circuit acting on a quantum register of three qubits
circ = QuantumCircuit(3)
# Add a H gate on qubit 0, putting this qubit in superposition.
circ.h(0)
# Add a CX (CNOT) gate on control qubit 0 and target qubit 1, putting
# the qubits in a Bell state.
circ.cx(0, 1)
# Add a CX (CNOT) gate on control qubit 0 and target qubit 2, putting
# the qubits in a GHZ state.
circ.cx(0, 2)

circ.draw('mpl', style='clifford')
```

### Qiskit Runtime : Sampler & Estimator

Qiskit IBM Runtime is a new environment offered by IBM Quantum that streamlines quantum computations and provides optimal implementations of the Qiskit primitives sampler and estimator for IBM Quantum hardware.

It is designed to use additional classical compute resources to execute quantum circuits with more efficiency on quantum processors, by including near-time computations such as error suppression and error mitigation. Examples of error suppression include dynamical decoupling, noise-aware compilation, error mitigation including readout mitigation, zero-noise extrapolation (ZNE), and probabilistic error cancellation (PEC).

```python
qc_example = QuantumCircuit(3)
qc_example.h(0)          # generate superpostion
qc_example.h(1)          # generate superpostion
qc_example.p(np.pi/2,0)  # add quantum phase
qc_example.cx(0,1)       # 0th-qubit-Controlled-NOT gate on 1st qubit
qc_example.cx(0,2)       # 0th-qubit-Controlled-NOT gate on 2nd qubit
qc_example.draw('mpl', style='clifford')
```

### Parameterized circuits

### Composite Gates: Build a sub-circuit with to_instruction() method:

This is useful when you want to reuse a certain quantum circuit multiple times in your main quantum circuit

```python
# Build a sub-circuit
sub_q = QuantumRegister(2)
sub_circ = QuantumCircuit(sub_q, name='sub_circ')
sub_circ.h(sub_q[0])
sub_circ.crz(1, sub_q[0], sub_q[1])
sub_circ.barrier()
sub_circ.id(sub_q[1])
sub_circ.u(1, 2, -2, sub_q[0])
sub_circ.draw('mpl', style = 'clifford')
```

```python
# Convert to a gate and stick ( append() ) it into an arbitrary place in the bigger circuit
sub_inst = sub_circ.to_instruction()
```

```python
# Create a main QuantumCircuit object
# Add the sub-circuit instruction to the main circuit with the append method

qr = QuantumRegister(4, 'q')
circ = QuantumCircuit(qr)
circ.h(qr[3])
circ.cx(qr[3], qr[0])
circ.cx(qr[1], qr[2])
circ.append(sub_inst, [qr[1], qr[2]])
circ.draw('mpl', style = 'clifford')
```

```python
# When desired, or before compilation, sub-circuits will be decomposed via the decompose method.
circ.decompose().draw('mpl')
```

### The Qiskit circuit library :
https://medium.com/qiskit/a-guide-to-the-qiskit-circuit-library-36ee0f189956

A collection of valuable, well-studied circuits and gates that Qiskit users can easily plug into their quantum circuits. These include useful tools for a wide array of quantum computing applications—including quantum machine learning, quantum chemistry, circuits for benchmarking, and even a number of circuits that are hard to simulate classically.

The quantum Fourier transform (QFT) is an essential building block in a number of quantum algorithms including quantum phase estimation and the famous Shor’s algorithm.

```python
def build_qft(n):
   circuit = QuantumCircuit(n)
   for j in reversed(range(n)):
       circuit.h(j)
       for k in reversed(range(j)):
           circuit.cp(np.pi * 2. ** (k - j), j, k)

   for j in range(n // 2):
       circuit.swap(j, n - j - 1)

   return circuit
```

```python
qft = build_qft(5)
qft.draw('mpl', style = 'clifford' )
```

### plot_bloch_vector

The plot_bloch_vector function in Qiskit is used to plot a 3D Bloch sphere and a vector on it. The Bloch sphere is a representation used in quantum computing to visualize the state of a single qubit — the fundamental unit of quantum information.
